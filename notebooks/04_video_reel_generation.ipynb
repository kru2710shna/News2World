# Video Reel Storyboard Generation

Takes a structured summary and creates a TikTok/Reels-style storyboard with scene breakdowns, captions, and image prompts.

```python
import json
from dataclasses import dataclass, asdict

@dataclass
class Scene:
    id: int
    timestamp: str
    caption: str
    image_prompt: str
    duration_sec: int = 3

def generate_storyboard(highlights: list[str], total_duration: int = 12) -> list[Scene]:
    """Generate a storyboard from key highlights."""
    n = len(highlights)
    per_scene = total_duration // n
    scenes = []
    for i, h in enumerate(highlights):
        start = i * per_scene
        end = start + per_scene
        scenes.append(Scene(
            id=i + 1,
            timestamp=f'0:{start:02d} – 0:{end:02d}',
            caption=h,
            image_prompt=f'News illustration: {h}',
            duration_sec=per_scene,
        ))
    return scenes
```

```python
highlights = [
    'Breaking: Ceasefire deal reached after months of negotiations in Doha.',
    'Hamas agrees to release hostages in phased exchange over 60 days.',
    'Humanitarian convoys begin entering Gaza within 48 hours of the deal.',
    'World leaders react — hope for lasting peace, but challenges remain.',
]

storyboard = generate_storyboard(highlights)
print(f'Generated {len(storyboard)} scenes:')
for s in storyboard:
    print(f'  Scene {s.id} ({s.timestamp}): {s.caption}')
```
